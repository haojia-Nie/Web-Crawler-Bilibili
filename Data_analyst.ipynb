```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

## gender

```python
data = pd.read_csv('set_1.csv')
gender_col = data.gender

unknown = 0
male = 0
female = 0

for row in gender_col:
    if row == '保密':
        unknown += 1
    elif row == '男':
        male += 1
    elif row == '女':
        female += 1

        
        
sizes = [unknown, male, female]
labels = 'unknown', 'male', 'female'
colors = ['orange', 'LightBlue', 'pink']
plt.pie(sizes, labels = labels, colors = colors,  autopct = '%1.1f%%', shadow = True)
plt.axis('equal')
plt.title('Gender pie chart')


plt.show()

```

## level and Vip 

```python
level_col = data.level
vip_col = data.vip


zero_one = 0
two_three = 0
four_five = 0
five_up = 0

zero_one_vip = 0
two_three_vip = 0
four_five_vip = 0
five_up_vip = 0

def if_vip(num):
    return vip_col[num]

for row in level_col:
    if  0<= row <= 1:
        if if_vip(row):
            zero_one_vip += 1
        else:
            zero_one += 1
        
    elif 2 <= row <= 3:
        if if_vip(row):
            two_three_vip += 1
        else:
            two_three += 1
            
    elif 4 <= row <= 5 :
        if if_vip(row):
            four_five_vip += 1
        else:
            four_five += 1

    elif row > 5:
        if if_vip(row):
            five_up_vip += 1
        else:
            five_up += 1
            
menStd = (20, 0, 25, 5)
womenStd = (10, 0, 20, 5)        

vip = (zero_one_vip, two_three_vip, four_five_vip, five_up_vip)
non_vip = (zero_one, two_three, four_five, five_up)  
       
x = np.arange(4)  
width = 0.35

p1 = plt.bar(x, vip, width, yerr = womenStd)
p2 = plt.bar(x, non_vip, width, bottom = vip, yerr = menStd)

plt.ylabel('users number')
plt.xlabel('level of users')
plt.yticks(np.arange(0, 500, 100))
plt.xticks(x, ('0 to 1', '2 to 3', '4 to 5', 'above 5'))

plt.title('Level of users in relation to vips')
plt.legend((p1[0], p2[0]), ('vip', 'not vip'))


plt.show()
```

## following and follower

```python
# following

data = pd.read_csv('set_1.csv')
follower_col = data.follower

level1 = 0 # below 10 followers
level2 = 0 # 10 to 999 followers
level3 = 0 # 1000 to 9999 followers
level4 = 0 # more than 10000 followers


for row in follower_col:
    if row < 10:
        level1 += 1
    elif row < 999:
        level2 += 1
    elif row <= 9999:
        level3 += 1
    else:
        level4 += 1

        
        
sizes = [level1, level3, level2, level4]
labels = 'below 10', '1000 to 9999', '10 to 999', 'above 10000'
colors = ['lightblue', 'DarkRed', 'lightyellow', 'black']
plt.pie(sizes, labels = labels, colors = colors,  autopct = '%1.1f%%', shadow = True)
plt.axis('equal')
plt.title('Follower pie chart')


plt.show()
```

```python
# the most popular user has 11168 follower, followed by 6759 and 5066 respectively

pd.Series.sort_values(follower_col, ascending = False)
```

```python
# following pie chart: very similar 

data = pd.read_csv('set_1.csv')
following_col = data.following


level1 = 0 # below 10 following
level2 = 0 # 10 to 99 following
level3 = 0 # 100 to 999 following
level4 = 0 # more than 1000 following


for row in following_col:
    if row < 10:
        level1 += 1
    elif row < 100:
        level2 += 1
    elif row <= 999:
        level3 += 1
    else:
        level4 += 1

        
        
sizes = [level1, level3, level2, level4]
labels = 'below 10', '10 to 99', '100 to 999', 'above 1000'
colors = ['lightblue', 'DarkRed', 'lightyellow', 'black']
plt.pie(sizes, labels = labels, colors = colors,  autopct = '%1.1f%%', shadow = True)
plt.axis('equal')
plt.title('Following pie chart')


plt.show()
```
